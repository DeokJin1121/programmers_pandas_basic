## Pandas참조 후 titanic 데이터를 가져오기
- name열이 index로 설정

```python
import pandas as pd
titanic = pd.read_excel('titanic3.xls', index_col='name')
titanic
```

## 데이터프레임의 columns만 조회

```python
titanic.columns
```

## .drop()을 이용해서 2개의 열을 삭제
- home.dest
- embarked
- sibsp
- 원본에 반영

```python
titanic.drop(columns=['home.dest', 'embarked', 'sibsp'], inplace=True)
titanic
```

## del을 이용한 열 삭제
- parch

```python
del titanic['parch']
titanic.head()
```

## 사망자 명단 조회하기

```python
survived_0_mask = titanic['survived'] == 0
titanic[survived_0_mask]
```

## 시신 식별번호(body)가 없는 데이터 조회하기

```python
body_none_mask = titanic['body'].isnull()
titanic[body_none_mask]
```

살아있는 사람은 시신 식별번호가 없는 것은 당연

## 사망자이나 시신식별번호가 없는 명단 찾기

```python
titanic[survived_0_mask & body_none_mask]
```

사망자가 809명인데, 식별번호가 없는 명단이 688명이나 된다.

## 사망자이나 시신식별번호가 없는 명단으로 새로운 df 만들기
- `notfound_list`

```python
notfound_list = titanic[survived_0_mask & body_none_mask]
notfound_list
```

## notfound_list의 승객 등급별 개수 확인

```python
notfound_list['pclass'].value_counts()
```

3등급(하류층)이 많으나, 이건 당연할 수 있다. 비율을 구해보자.

## 전체 titanic 전체 승객 등급별 개수 확인

```python
titanic['pclass'].value_counts()
```

## 승객 등급별 실종 비율 확인

```python
notfound_list['pclass'].value_counts() / titanic['pclass'].value_counts()
```

## 1. 사망자의 하류층 티겟 가격 분포와 2. 생존자의 하류층 티켓 가격 분포를 확인해보자.

```python
dead_mask = titanic['survived'] == 0
alive_mask = titanic['survived'] == 1
pclass_3_mask = titanic['pclass'] == 3
```

```python
# 사망자 3등급 df 생성
p3_dead = titanic[pclass_3_mask & dead_mask]
```

```python
# 생존자 3등급 df 생성
p3_alive = titanic[pclass_3_mask & alive_mask]
```

```python
# 사망자 3등급 기술 통꼐 환인
p3_dead.describe()
```

```python
# 생존자 3등급 기술 통계 확인
p3_alive.describe()
```

3등급에서 티켓가격의 차이가 생존유무와 관련성이 있어보이지 않는다.
