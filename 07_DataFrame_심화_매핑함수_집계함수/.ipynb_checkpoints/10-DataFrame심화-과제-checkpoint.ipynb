## 수익을 다음으로 분류
- 60000이상 고임금
- 30000이상~60000미만 평균임금
- 30000미만 저임금
- 열 이름은 'Class_Income'

```python
# apply, apply_map
def class_income(Class_Income):
    if Class_Income >= 60000:
        return '고임금'
    elif 30000 <= Class_Income < 60000:
        return '평균임금'
    elif Class_Income < 30000:
        return '저임금'
    return None
```

```python
customer['Class_Income'] = customer['Income'].apply(class_income)
customer
```

```python
temp = customer.pop('Class_Income')
customer.insert(3, 'Class_Income', temp)
customer
```

## Alone, Absurd, YOLO를 Single로 변경
1. reset_index()로 되돌리고
2. replace으로 값을 바꾼다음에
3. 다시 set_index()로 설정해보자.
4. 실제로 잘 변경되었는지 get_level_values(0)후 value_counts()로 데이터 경우의 수를 확인해보자.

```python
customer.head()
```

```python
# 1. reset_index()
customer.reset_index(inplace=True)
customer
```

```python
# 2. replace, map으로 값을 바꾼 다음에
customer['Marital_Status'].replace({'Alone': 'Single', 'Absurd': 'Single', 'YOLO': 'Single'}, inplace=True)
customer
```

```python
customer['Marital_Status'] = customer['Marital_Status'].replace({'Alone': 'Single', 'Absurd': 'Single', 'YOLO': 'Single'})
customer
```

```python
# 3. 다시 set_index()로 설정
customer.set_index(['Marital_Status', 'Education'], inplace=True)
customer.sort_index(inplace=True)
customer
```

```python
# 실제로 잘 변경되었는지 get_level_values(0)후 value_counts()로 데이터 경우의 수를 확인해보자
customer.index.get_level_values(0).value_counts()
```
